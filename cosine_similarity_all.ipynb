```python
# load package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# read article 3 and 415
df_all = pd.read_csv('../input/stage1_unigram.csv', header = 0, names=["id", "article", "words"])
# df_all = pd.read_csv('../input/stage2_unigram.csv', header = 0, names=["id", "article", "words"])
# df_all = df_all[0:5]
df_all
```

```python
# compute tf
vectorizer = CountVectorizer() 
text_count_vector = vectorizer.fit_transform(df_all.words)
tf_vector = text_count_vector.toarray()
tf_vector
```

```python
# compute tf-idf
tfidf_transfomer = TfidfTransformer()
docs_tfidf = tfidf_transfomer.fit_transform(text_count_vector)
tfidf_vector = docs_tfidf.toarray()
tfidf_vector
```

```python
tf_similarity = cosine_similarity(tf_vector, tf_vector)
tf_similarity
```

```python
tfidf_similarity = cosine_similarity(tfidf_vector, tfidf_vector)
tfidf_similarity
```

```python
result_df = pd.DataFrame({'Test':[], 'Reference': [], 'TF-CosineSimilarity': [], 'TFIDF-CosineSimilarity': []})

for i in range(len(tf_vector)):
    for j in range(len(tf_vector)):
        if i == j:
            continue
        dict = {'Test': str(df_all["id"][i]), 'Reference': str(df_all["id"][j]),
                'TF-CosineSimilarity': tf_similarity[i][j], 
                'TFIDF-CosineSimilarity': tfidf_similarity[i][j]}
        result_df = result_df.append(dict, ignore_index=True)
```

```python
result_df
```

```python
result_df.to_csv("../output/cosine_similarity_results_stage1.csv", index= False)
# result_df.to_csv("../output/cosine_similarity_results_stage2.csv", index= False)
```
